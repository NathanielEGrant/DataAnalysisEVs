## Background & Prior Work


Internal combustion engines (ICE) vehicles emit more pollutants than just the exhaust that exits from their tailpipes as well as through fuel system evaporation and during fueling.
Gasoline, emissions are created during the extraction of petroleum from the earth, refinement, distribution to fuel stations, and combustion in cars. These emissions are also known as upstream emissions or well-to-wheel emissions.
 Modern ICE manufacturers have reduced their CO2 emissions; however the production process still has a detrimental effect on the environment.

Direct emissions from all-electric vehicles are zero. Emissions in Electric Vehicles are produced by electric power plants, and more are produced during the extraction, processing, and distribution of the major energy sources that go into making the electricity.
 There are upstream emissions produced during the manufacture of EV batteries. In reality, EV manufacture can be more environmentally damaging than ICE production. EVs are still the cleanest form of transportation because their entire life cycle is significantly more environmentally friendly.
Many businesses have started using electric vehicles (EVs) as a result of rising fuel prices and a push for greener initiatives known for their high fuel efficiency, EVs might be an affordable option to lower running costs. An electric car often costs less per mile than a gasoline-powered vehicle, at around 3 cents as opposed to 10 cents.
 Electric Vehicles are a more environmentally friendly option than gas or diesel automobiles in addition to having cheaper fuel expenses. They can lower carbon gas emissions by getting rid of exhaust. This benefit aids companies in continuing to operate sustainably and in compliance with legal requirements.

Electric motors in Electric Vehicles are powered by rechargeable lithium-ion batteries, which are also used to power computers and smartphones. Electric Vehicles can be charged by 
 plugging into external power sources just like electrical devices do. In addition to being cleaner than internal combustion engines, lithium-ion batteries frequently perform better. Many have an 8â€“10 year guaranteed life span. 

Many electric charging stations are provided for the EV batteries just like gas stations. An EV's battery life determines how far it can go between charging points. Electric Vehicles do not emit Carbon dioxide which helps reduce the pollution in the air. 
There are three different Kinds of Electric vehicles.

- Hybrid electric vehicles (HEVs): Both gasoline and electricity are used to power this car as well but They are distinct from PHEVs because they exclusively use regenerative braking to recharge their batteries. When driving, these EVs use their electric motor; but, as the load or speed increases, the electric motor is supplemented by the gasoline engine.

# Setup and Data Cleaning

The code below reads a CSV file called "Electric_Cars_VS_Motor_Cars.csv" using pandas and separates the fields using semicolons as the separator. The resulting dataframe is printed to the console.

Then, a new dataframe called newTable is created by dropping four columns from the original dataframe (EV Stock BEV, EV Stock PHEV, Electricity demand BEV, and Electricity demand BHEV). These columns were decided to be uneccessary to the data analysis that we wanted to do as a group. The resulting dataframe is also printed to the console.

The "Oil displacement (Million barrels per day)" column has data that is interpreted by python as a string. Inorder to change the data types to float we had to used the .strreplace method to change the commas "," to periods "." in order to be able to undergo the next functionality in te line under where we typecasted the whole column from strings to floats.

The original dataframe has data on electric and motor car sales, EV stock and electricity demand, oil displacement, and CO2 emissions for BEV and PHEV, among other information. The new dataframe only includes information on EV sales, oil displacement, and CO2 emissions. The units for some of the values are given in the comments.

```python
import pandas as pd
url = "Electric_Cars_VS_Motor_Cars.csv"
df = pd.read_csv(url, sep = ';')
df
```

```python
# drop unused columns of the table: 'EV Stock BEV', 'EV Stock PHEV', 'Electricity demand BEV', 'Electricity demand BHEV'
newTable = df.drop(columns = ['EV Stock BEV', 'EV Stock PHEV', 'Electricity demand BEV', 'Electricity demand BHEV'])

# replace the comma with a period in the "Oil displacement (Million barrels per day)" column
newTable["Oil displacement (Million barrels per day)"] = newTable["Oil displacement (Million barrels per day)"].str.replace(",", ".")

# convert the "Oil displacement (Million barrels per day)" column to float and round to the nearest integer
newTable["Oil displacement (Million barrels per day)"] = newTable["Oil displacement (Million barrels per day)"].astype(float)

newTable
```

This table shows the EV car sales and Average CO2 emissions per year from 2010 to 2021.

### Test 1 - Correlation Analysis

To conduct a correlation analysis using Python code, we first need to import the necessary libraries:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = newTable
```

Next, we can calculate the correlation coefficient between the two variables of interest (EV sales and oil displacement) using the corr() function in pandas:

```python
correlation_coefficient = df['EV sales BEV'].corr(df['Oil displacement (Million barrels per day)'])
```

Now lets print the correlation coefficient:

```python
print(correlation_coefficient)
```

The output will be a value between -1 and 1, where a negative value indicates a negative correlation, and a positive value indicates a positive correlation. A value of 0 indicates no correlation.

To visualize the correlation, we can create a scatter plot using seaborn:

```python
sns.scatterplot(x='EV sales BEV', y='Oil displacement (Million barrels per day)', data=df)
plt.show()
```

 If there is a negative correlation, we should see a downward trend in the data points on the scatter plot.

If we want to see the correlation between multiple variables, we can create a correlation matrix using seaborn:

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

This will create a heatmap with the correlation coefficients between all pairs of variables in the dataframe.

### Test 1 - Conclusions

A correlation coefficient of .964 suggests a very strong positive correlation between the production of electric vehicles and the amount of oil displacement (in million barrels per day) in the United States. This means that as the production of electric vehicles increases, the amount of oil displacement also tends to increase.

It is important to note that correlation does not necessarily imply causation. While the strong positive correlation suggests that there may be a relationship between the two variables, further research is needed to establish causality and to determine whether the production of electric vehicles is indeed leading to a reduction in air pollution through a decrease in oil displacement.

### Test 2 - Regression Analysis

To perform a regression analysis, we can use the statsmodels library in Python. Here are the steps to perform a multiple linear regression analysis:

```python
#import required libraries
import statsmodels.api as sm
```

Next, we need to define the independent and dependent variables:

```python
X = df[['EV sales BEV', 'EV sales PHEV', 'Total Motor cars production per year']]
y = df['Oil displacement (Million barrels per day)']
```

Here, we are using three independent variables - the number of BEV sales, the number of PHEV sales, and the total motor car production per year. We are using oil displacement as the dependent variable.

```python
# add the constant term to the indep variables

X = sm.add_constant(X)

# and fit the mult linear regression model

model = sm.OLS(y, X).fit()
```

In the last step we built the multiple linear regression model and now we need to print it. We will do this by using the .summary() functionality.

This will output a summary table of the regression analysis, which includes information such as the coefficients, standard errors, t-values, and p-values for each independent variable.

```python
print(model.summary())
```

### Test 2 - Conclusions

Based on the given regression analysis results, we can make the following conclusions:




1. The model is statistically significant as the F-statistic has a low p-value of 3.22e-05, indicating that the independent variables are jointly significant in explaining the dependent variable.

2. The R-squared value of 0.939 indicates that the independent variables in the model explain 93.9% of the variation in the dependent variable, which is a high percentage.
- However, the adjusted R-squared value of 0.917 suggests that the model probably has some overfitting issues.

3. The coefficients of the independent variables are all very small, with the highest being 1.237e-07. This suggests that the impact of electric vehicle production on oil displacement is negligible.

4. None of the independent variables are statistically significant as their p-values are all greater than 0.05, indicating that they do not have a significant effect on the dependent variable.

In summary, the regression analysis suggests that there is no significant impact of electric vehicle production on oil displacement in the United States after controlling for other variables. The model may have some overfitting issues, which could limit its usefulness in making accurate predictions.
